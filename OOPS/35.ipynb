```python
class product:
    def __init__(self, product_id: int, name: str, price: float):
        self.product_id = name
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (${self.price:.2f})"

class cart:
    def __init__(self):
        self.cart_items = {}

    def add_product(self, product: product, quantity: int = 1):
        if product.product_id in self.cart_items:
            self.cart_items[product.product_id]['quantity'] += quantity
        else:
            self.cart_items[product.product_id] = {
                "product": product,
                "quantity": quantity
            }
    
    def remove_product(self, product_id: int):
        if product_id in self.cart_items:
            del self.cart_items[product_id]
        else:
            print("Product not found in cart")
        
    def update_quantity(self, product_id: int, quantity: int):
        if product_id in self.cart_items:
            if quantity <= 0:
                self.remove_product(product_id)
            else:
                print("Product not found in cart.")

    def calculate_total(self):
        return sum(
            item['product'].price * item['quantity']
            for item in self.cart_items.values()
        )

    def display_cart(self):
        if not self.cart_items:
            print("your cart is empty")
            return

        print("cart items: ")

        for item in self.cart_items.values():
            product = item['product']
            quantity = item['quantity']

            print(f" - {product.name} (x{quantity})@ ${product.price:.2f} = $ {product.price * quantity:.2f}")

        print(f"Total: ${self.calculate_total():.2f}")

laptop = product(101, "Laptop", 59999.00)
headphones = product(102, "headphones", 2999.00)
mouse = product(103, "Mouse", 1499.20)

mycart = cart()

mycart.add_product(laptop)
mycart.add_product(headphones, 2)
mycart.add_product(mouse, 1)

mycart.display_cart()

mycart.update_quantity(102, 1)
mycart.remove_product(103)

print("\n After updates")

mycart.display_cart()
```
